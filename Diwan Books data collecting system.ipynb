Scrap all the books in the following website and
save it in a csv file with the next formating
['book name','price','rate','category']
price==> float
rate==>int

http://books.toscrape.com/index.html

```python
ul='http://books.toscrape.com/catalogue/category/books/mystery_3/page-1.html'
re=requests.get(ul)
re.text
```

```python
soup=BeautifulSoup(re.text,'lxml')
soup
```

```python
file_name=soup.find('div',attrs={'class':'page-header action'}).find('h1').get_text()
file_name
```

```python
name_book = soup.find_all('article',attrs={'class':'product_pod'})[0].find('h3').find('a').get_text()
name_book
```

```python
price=soup.find_all('div',attrs={'class':'product_price'})[0].find('p',attrs={'class':'price_color'}).get_text()
price= float(price[2:])
price
```

```python
rates={'one':1,'two':2,'three':3,'four':4,'five':5}
r=soup.find_all('article',attrs={'class':'product_pod'})[0].find('p',attrs={'class':'star-rating'}).get('class')[1]
r
```

```python
for i in soup.find_all('aside',attrs={'class':'sidebar col-sm-4 col-md-3'}):
  category=i.find('div',attrs={'class':'side_categories'}).get_text()
category
soup.find_all('article',attrs={'class':'product_pod'})
```

```python
rates={'one':1,'two':2,'three':3,'four':4,'five':5}
with open(file_name,'w',encoding='utf-8',newline='')as file:
    writer=csv.writer(file)
    writer.writerow(['Name_Book','Price','Rate','Category'])
    for i in soup.find_all('aside',attrs={'class':'sidebar col-sm-4 col-md-3'}):
      category=i.find('div',attrs={'class':'side_categories'}).get_text().split()
      for i in soup.find_all('article',attrs={'class':'product_pod'}):
         book_name=i.find('h3').find('a').get_text()
         price=float(i.find('p',attrs={'class':'price_color'}).get_text()[2:])
         rate=i.find('p',attrs={'class':'star-rating'}).get('class')[1]
         writer.writerow([book_name,price,rate,category])
```
